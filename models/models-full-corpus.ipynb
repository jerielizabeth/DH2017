```python
import gensim
import os
import pyLDAvis.gensim
```

```python
pyLDAvis.enable_notebook()
```

```python
base_dir = "/Users/jeriwieringa/Dissertation/drafts/data/module-3/"
scheme = "full_corpus"
```

```python
print('Period {}'.format(scheme))

corpus = gensim.corpora.MmCorpus(os.path.join(base_dir, "corpora", scheme, "{}.mm".format(scheme)))
dictionary = gensim.corpora.Dictionary.load(os.path.join(base_dir, "corpora", scheme, "{}.dict".format(scheme)))
model = gensim.models.LdaModel.load(os.path.join(base_dir, "models/final/{}.model".format(scheme, scheme)))

pyLDAvis.gensim.prepare(model, corpus, dictionary)
```
