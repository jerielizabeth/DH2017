```python
import gensim
import os
import pyLDAvis.gensim
```

```python
pyLDAvis.enable_notebook()
```

```python
base_dir = "/Users/jeriwieringa/Dissertation/drafts/data/module-3/"
scheme = "historical_periods"
```

```python
historical_periods = ['1849-to-1858',
                      '1859-to-1875',
                      '1876-to-1896',
                      '1897-to-1909',
                      '1910-to-1920'
                     ]
```

```python
print('Period {}'.format(historical_periods[0]))

corpus = gensim.corpora.MmCorpus(os.path.join(base_dir, "corpora", scheme, "{}.mm".format(historical_periods[0])))
dictionary = gensim.corpora.Dictionary.load(os.path.join(base_dir, "corpora", scheme, "{}.dict".format(historical_periods[0])))
model = gensim.models.LdaModel.load(os.path.join(base_dir, "models/final/historical-period-{}.model".format(historical_periods[0])))

pyLDAvis.gensim.prepare(model, corpus, dictionary)
```
